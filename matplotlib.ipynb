```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 25)
y = (x * x) + 2
```

```python
x, y
```

```python
fig = plt.figure()

# left, bottom, width, height
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.15, 0.45, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes2.plot(y, x, 'b')
```

```python
fig = plt.figure(figsize=(16, 9), dpi=350)
plt.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
for ax in axes:
  ax.plot(x, y)
```

```python
fig, axes = plt.subplots()
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Graph")
axes.plot(x**2)
axes.plot(x**3)
axes.legend(["y = x **2", "y = x **3"], loc = 2)
```

```python
fig, axes = plt.subplots()
axes.plot(x+3, color = 'red', alpha = .6)
axes.plot(x+4, linewidth = 4)
axes.plot(x+5, linestyle=':', marker="o", markersize = 6, markerfacecolor = "black")
```

```python
fig, axes = plt.subplots(1, 2, figsize = (10, 6), dpi = 200)
axes[0].plot(x, x**2, x, x**3, lw = 2)
axes[0].grid(True) # Showing grid in the chart

# Setting limit for the chart
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].set_ylim([0, 100])
axes[1].set_xlim([0, 5])
```

# Other 2D graphs

```python
n = np.linspace(1, 10, 10)
m = np.linspace(0, 4, 10)
```

```python
fig, axes = plt.subplots(1, 4, figsize = (14, 5))
axes[0].set_title("Scatter plot")
axes[0].scatter(m, n*n)

axes[1].set_title("Step plot")
axes[1].step(m, n**m, lw = 2)

axes[2].set_title("Bar plot")
axes[2].bar(n, n**2, align = "center", width = 0.8)

axes[3].set_title("Fill between")
axes[3].fill_between(n, n**2, m**2)
```

# Contour plot

```python
# matplot colormap
import matplotlib.cm as cm
```

```python
delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2
```

```python
fig, ax = plt.subplots(figsize = (14, 7))
CS = ax.contour(X, Y, Z, cmap = cm.jet,)
ax.clabel(CS, inline = 1, fontsize = 10)
```

# 3D plot

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
colo = colo = np.random.randn(10, 1000)*1000
fig = plt.figure(figsize = (14, 7), dpi = 250)
ax = fig.add_subplot(1, 1, 1, projection = '3d')
color_map = cm.ScalarMappable(cmap=cm.gray)
color_map.set_array(colo)
ax.plot_surface(X, Y, Z, rstride = 4, cstride = 4)
plt.colorbar(color_map)
```

# Pie Chart

```python
languages = ['Python', 'Javascript', 'C++', 'PHP', 'Go', 'Rust']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['red', 'blue', 'green', 'white', 'orange', 'purple']
```

```python
explode = (0.1, 0, 0, 0, 0, 0)
plt.pie(popularity, explode=explode, labels=languages, colors=colors, autopct='%1.1f%%', shadow=True, startangle=180)
```
